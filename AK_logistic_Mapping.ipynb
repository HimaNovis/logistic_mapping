```
import numpy as np
import matplotlib.pyplot as plt
```

```
n=100
x=0.5
r=0.8
a=np.zeros(n)
i=0
while(i<n):
    a[i]=r*x*(1-x)
    x=r*x*(1-x)
    i=i+1
t=np.linspace(0,n-1,n)
plt.plot(t,a)
```
